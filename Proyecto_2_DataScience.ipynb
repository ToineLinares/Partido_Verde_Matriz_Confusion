```python
Y_v18 = np.ones(len(verde_18), dtype = np.int)
Y_g18 = np.ones(len(verde_18), dtype = np.int)
Y_v22 = np.ones(len(verde_18), dtype = np.int)
Y_g22 = np.ones(len(verde_18), dtype = np.int)
```

```python
Y_v18[above_mean_g18] = 1
Y_v18[below_mean_g18] = 0

Y_v22[above_mean_g22] = 1
Y_v22[below_mean_g22] = 0


X_v18 = np.zeros((len(verde_18), len(verde_18)))
X_v18[:,0] = verde_18
for i in range(len(verde_18)):
    for j in range(1, len(verde_18)):
        X_v18[i,j] = i + 1
X_v18
```
